```python
#this script is made to visualize the data from the csv file using matplotlib and seaborn

import pandas as pd, matplotlib.pyplot as plt, numpy as np
import ast, seaborn as sns
from adjustText import adjust_text
from datetime import datetime as dt
from matplotlib.ticker import FuncFormatter

csv = pd.read_csv("csv_files/top_paying_jobs.csv")
```

```python
csv['job_posted_date'] = pd.to_datetime(csv['job_posted_date'])
```

```python
csv.info()
```

```python
#Picture 1
csv.plot(kind='barh', x='job_title', y='salary_year_avg')
plt.gca().invert_yaxis()
plt.ylabel("")
plt.xlabel("Average Salary (USD)")
plt.title("Top Paying Remote Data Science Jobs")
plt.legend().remove()
plt.gca().figure.set_dpi(1500)
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
```

```python
csv = pd.read_csv("csv_files/top_paying_jobs_skills.csv")
```

```python
csv.info()
```

```python
csv_plot = csv.pivot_table(index='skills', values='salary_year_avg').sort_values(by='salary_year_avg', ascending=False).reset_index()
sns.barplot(data=csv_plot, x='salary_year_avg', y='skills', palette="viridis", errorbar=None)
plt.ylabel("")
plt.xlabel("Average Salary (USD)")
plt.title("Skills Associated With Highest Paid Jobs")
plt.legend().remove()
plt.gca().figure.set_dpi(1500)
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
```
